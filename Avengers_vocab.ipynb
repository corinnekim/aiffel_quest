```python
def sen_idx () :
    import re
    from collections import Counter

    txt_data = open('/content/Avengers.txt', 'r', encoding = 'utf-8-sig')
    txt = txt_data.read()

    txt = txt.lower() #모두 소문자
    txt = re.sub(r"'", '', txt) #' 지우기
    txt = re.sub(r'[^a-z0-9\s]', ' ', txt) #특수기호 제거
    txt = re.sub(r'\n', ' ', txt) #줄바꿈은 한칸으로
    txt = re.sub(r'\s+', ' ', txt) #공백 뭉치 1개로
    txt = txt.strip() #맨앞 맨뒤 공백 삭제

    txt_split = txt.split(' ')
    word_count = Counter(txt_split)
    word_count = word_count.most_common()

    output = dict()
    for index, (word, count) in enumerate(word_count) :
        output[word] = index

    sen = input('영어로 입력해주세요')
    sen_split = sen.split()

    result_idx = []
    for i in sen_split :
        result_idx.append(output[i])

    txt_data.close()
    return result_idx
```
